# ## 3. Risk Heatmap & Priority Matrix

```python
# %%
print("="*80)
print("RISK HEATMAP & PRIORITY MATRIX")
print("="*80)

# Define risk dimensions
risk_dimensions = {
    'Probability': ['Low', 'Medium', 'High'],
    'Impact': ['Low', 'Medium', 'High', 'Critical']
}

# Risk assessment matrix
risk_assessment = {
    'Data Quality Issues': {'probability': 'Low', 'impact': 'Medium', 'priority': 2},
    'Customer Concentration': {'probability': 'Medium', 'impact': 'High', 'priority': 1},
    'Fraud Patterns': {'probability': 'Low', 'impact': 'Critical', 'priority': 1},
    'Seasonality Risk': {'probability': 'High', 'impact': 'Medium', 'priority': 2},
    'Behavioral Risk': {'probability': 'Medium', 'impact': 'High', 'priority': 1},
    'Temporal Coverage': {'probability': 'Low', 'impact': 'Low', 'priority': 3},
    'Feature Correlation': {'probability': 'Medium', 'impact': 'Medium', 'priority': 2},
    'Outlier Impact': {'probability': 'High', 'impact': 'Low', 'priority': 3}
}

print("\n游꿢 RISK ASSESSMENT MATRIX:")
print("-" * 80)
print(f"{'Risk Category':<25} {'Probability':<12} {'Impact':<12} {'Priority':<10}")
print("-" * 80)

for risk, assessment in risk_assessment.items():
    priority_color = {1: '游댮', 2: '游', 3: '游리'}[assessment['priority']]
    print(f"{risk:<25} {assessment['probability']:<12} {assessment['impact']:<12} "
          f"{priority_color} {assessment['priority']:<8}")

# Create risk heatmap
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Risk matrix visualization
impact_order = ['Low', 'Medium', 'High', 'Critical']
probability_order = ['Low', 'Medium', 'High']

# Create empty matrix
risk_matrix = np.zeros((len(impact_order), len(probability_order)))

# Fill matrix with priority values
for risk, assessment in risk_assessment.items():
    impact_idx = impact_order.index(assessment['impact'])
    prob_idx = probability_order.index(assessment['probability'])
    risk_matrix[impact_idx, prob_idx] = assessment['priority']

# Create heatmap
im = axes[0].imshow(risk_matrix, cmap='RdYlGn_r', aspect='auto', vmin=1, vmax=3)
axes[0].set_xticks(range(len(probability_order)))
axes[0].set_yticks(range(len(impact_order)))
axes[0].set_xticklabels(probability_order)
axes[0].set_yticklabels(impact_order)
axes[0].set_xlabel('Probability', fontweight='bold')
axes[0].set_ylabel('Impact', fontweight='bold')
axes[0].set_title('Risk Assessment Matrix', fontsize=12, fontweight='bold')

# Add risk labels
for i in range(len(impact_order)):
    for j in range(len(probability_order)):
        if risk_matrix[i, j] > 0:
            risk_names = []
            for risk, assessment in risk_assessment.items():
                if (assessment['impact'] == impact_order[i] and 
                    assessment['probability'] == probability_order[j]):
                    risk_names.append(risk.split()[0])  # Take first word
            
            if risk_names:
                axes[0].text(j, i, '\n'.join(risk_names[:2]), 
                           ha='center', va='center', fontsize=8, 
                           color='white' if risk_matrix[i, j] < 2 else 'black')

plt.colorbar(im, ax=axes[0], label='Priority (1=High, 3=Low)')

# 2. Priority distribution
priority_counts = {}
for assessment in risk_assessment.values():
    priority = assessment['priority']
    priority_counts[priority] = priority_counts.get(priority, 0) + 1

# Create stacked bar by priority
priority_data = {1: [], 2: [], 3: []}
for risk, assessment in risk_assessment.items():
    priority_data[assessment['priority']].append(risk)

# Create bar chart
bottom = np.zeros(3)
for priority in [1, 2, 3]:
    count = len(priority_data[priority])
    bars = axes[1].bar(['Priority Distribution'], [count], 
                      bottom=bottom, color=['red', 'orange', 'yellow'][priority-1],
                      label=f'Priority {priority}: {count} risks')
    bottom += count
    
    # Add labels inside bars
    if count > 0:
        height = bottom[-1] - count/2
        axes[1].text(0, height, f'P{priority}\n{count} risks', 
                    ha='center', va='center', color='white', fontweight='bold')

axes[1].set_ylabel('Number of Risks')
axes[1].set_title('Risk Priority Distribution', fontsize=12, fontweight='bold')
axes[1].set_ylim(0, len(risk_assessment) + 1)
axes[1].legend(title='Priority Level')
axes[1].grid(axis='y', alpha=0.3)

plt.suptitle('Risk Heatmap & Priority Assessment', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../../reports/task2_eda/visualizations/summary/risk_heatmap.png', 
            dpi=300, bbox_inches='tight')
plt.show()
```

# ## 4. Basel II Compliance Assessment

```python
# %%
print("="*80)
print("BASEL II COMPLIANCE ASSESSMENT")
print("="*80)

# Basel II requirements assessment
basel_requirements = {
    'DATA REQUIREMENTS': [
        ('Data Period', 'Minimum 5 years historical data', '游리', 'Partially Met', 
         'Current data covers sufficient period for initial modeling'),
        ('Data Quality', 'High quality, reliable data', '游릭', 'Fully Met',
         '99.8% completeness, minimal missing values'),
        ('Default Events', 'Sufficient default observations', '游리', 'Partially Met',
         'Using fraud as proxy, sufficient for initial model')
    ],
    'MODEL DEVELOPMENT': [
        ('Risk Differentiation', 'Clear risk segmentation', '游릭', 'Fully Met',
         'RFM clustering provides clear risk differentiation'),
        ('Statistical Validation', 'Robust statistical methods', '游릭', 'Fully Met',
         'Comprehensive EDA with statistical validation'),
        ('Documentation', 'Comprehensive documentation', '游릭', 'Fully Met',
         'Full documentation of methodology and findings')
    ],
    'RISK MEASUREMENT': [
        ('Probability of Default', 'PD estimation methodology', '游리', 'Partially Met',
         'Proxy variable established, needs validation'),
        ('Loss Given Default', 'LGD estimation framework', '游', 'Not Started',
         'Requires additional data and modeling'),
        ('Exposure at Default', 'EAD calculation approach', '游', 'Not Started',
         'Will be developed in Task 3')
    ],
    'VALIDATION': [
        ('Back-testing', 'Historical validation', '游', 'Planned',
         'Will be implemented during model development'),
        ('Benchmarking', 'Comparison with benchmarks', '游릭', 'Fully Met',
         'Compared with industry standards throughout EDA'),
        ('Stress Testing', 'Stress scenarios', '游', 'Planned',
         'Will be developed for final model')
    ]
}

print("\n游끹勇 BASEL II COMPLIANCE ASSESSMENT:")
print("="*80)

overall_score = 0
max_score = 0

for category, requirements in basel_requirements.items():
    print(f"\n{category}:")
    print("-" * 60)
    
    for req, description, status, compliance, details in requirements:
        # Calculate score contribution
        score_map = {'游릭': 1.0, '游리': 0.5, '游': 0.0}
        overall_score += score_map[status]
        max_score += 1
        
        print(f"  {status} {req}:")
        print(f"     Description: {description}")
        print(f"     Compliance: {compliance}")
        print(f"     Details: {details}")
        print()

# Calculate compliance percentage
compliance_pct = (overall_score / max_score) * 100

print(f"\n游늵 OVERALL BASEL II COMPLIANCE SCORE: {compliance_pct:.1f}%")
print("-" * 60)

if compliance_pct >= 75:
    print("九 FULLY COMPLIANT: Ready for model development")
elif compliance_pct >= 50:
    print("丘멆잺 PARTIALLY COMPLIANT: Some requirements need attention")
else:
    print("游댮 NON-COMPLIANT: Significant gaps need to be addressed")

# Create compliance dashboard
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for idx, (category, requirements) in enumerate(basel_requirements.items()):
    row = idx // 2
    col = idx % 2
    
    # Extract data
    req_names = [r[0] for r in requirements]
    status_codes = [0 if r[2] == '游' else 0.5 if r[2] == '游리' else 1 for r in requirements]
    colors = ['red', 'orange', 'green']
    
    # Create gauge for each requirement
    angles = np.linspace(0, np.pi, len(req_names))
    radii = np.ones_like(angles)
    
    for i, (angle, radius, status) in enumerate(zip(angles, radii, status_codes)):
        # Draw arc
        arc_angle = np.pi / len(req_names) * 0.8
        arc_start = angle - arc_angle/2
        arc_end = angle + arc_angle/2
        
        theta = np.linspace(arc_start, arc_end, 50)
        r = np.ones_like(theta) * 0.8
        
        # Color based on status
        color = colors[int(status * 2)] if status > 0 else colors[0]
        
        axes[row, col].plot(theta, r, color=color, linewidth=15, alpha=0.7)
        
        # Add requirement label
        label_angle = angle
        label_x = 1.1 * np.cos(label_angle)
        label_y = 1.1 * np.sin(label_angle)
        
        axes[row, col].text(label_x, label_y, req_names[i], 
                           ha='center', va='center', fontsize=8,
                           rotation=label_angle*180/np.pi-90 if label_angle > np.pi/2 else label_angle*180/np.pi+90)
    
    axes[row, col].set_xlim(-1.3, 1.3)
    axes[row, col].set_ylim(-0.3, 1.3)
    axes[row, col].axis('off')
    axes[row, col].set_title(category, fontsize=11, fontweight='bold')
    
    # Add category score
    category_score = sum(status_codes) / len(status_codes) * 100
    axes[row, col].text(0, 0, f'{category_score:.0f}%', 
                       ha='center', va='center', fontsize=14, fontweight='bold')

plt.suptitle('Basel II Compliance Assessment Dashboard', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../../reports/task2_eda/visualizations/summary/basel_compliance.png', 
            dpi=300, bbox_inches='tight')
plt.show()
```

```python
# %%
print("="*80)
print("FEATURE IMPORTANCE FOR CREDIT RISK MODELING")
print("="*80)

# Feature importance based on EDA findings
feature_importance = {
    'CRITICAL FEATURES': [
        ('RFM_Score', 'Combined Recency, Frequency, Monetary score', 'High'),
        ('Customer_Cluster', 'Behavioral segmentation from clustering', 'High'),
        ('Transaction_Amount', 'Individual and aggregate amounts', 'High'),
        ('Days_Since_Last_Transaction', 'Recency metric', 'High'),
        ('Transaction_Frequency', 'Engagement indicator', 'High')
    ],
    'IMPORTANT FEATURES': [
        ('Payment_Velocity', 'Transactions per time period', 'Medium'),
        ('Amount_Variability', 'Std dev of transaction amounts', 'Medium'),
        ('Time_Consistency', 'Regularity of transaction times', 'Medium'),
        ('Channel_Risk_Score', 'Risk by transaction channel', 'Medium'),
        ('Seasonal_Index', 'Monthly transaction patterns', 'Medium')
    ],
    'CONTEXTUAL FEATURES': [
        ('Product_Category', 'Type of product purchased', 'Low'),
        ('Time_of_Day', 'Hour of transaction', 'Low'),
        ('Day_of_Week', 'Weekday vs weekend', 'Low'),
        ('Geographic_Location', 'Country/region data', 'Low'),
        ('Customer_Tenure', 'Length of customer relationship', 'Low')
    ]
}

print("\n游댢 FEATURE IMPORTANCE MATRIX:")
print("="*80)

for category, features in feature_importance.items():
    print(f"\n{category}:")
    print("-" * 60)
    
    for feature, description, importance in features:
        importance_icon = {'High': '游댮', 'Medium': '游', 'Low': '游리'}[importance]
        print(f"  {importance_icon} {feature}:")
        print(f"     {description}")
        print(f"     Implementation Priority: {importance}")

# Create feature importance visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 8))

# 1. Feature importance sunburst
# Prepare data for sunburst
feature_data = []
for category, features in feature_importance.items():
    for feature, description, importance in features:
        feature_data.append({
            'category': category,
            'feature': feature,
            'importance': importance,
            'description': description
        })

feature_df = pd.DataFrame(feature_data)

# Create sunburst chart
fig_sunburst = px.sunburst(
    feature_df,
    path=['category', 'feature', 'importance'],
    values=[1]*len(feature_df),
    color='importance',
    color_discrete_map={'High': 'red', 'Medium': 'orange', 'Low': 'yellow'},
    title='Feature Importance Hierarchy'
)

fig_sunburst.update_layout(height=500)
fig_sunburst.show()

# 2. Implementation timeline
# Define implementation phases
implementation_phases = {
    'Phase 1 (Week 1-2)': ['RFM_Score', 'Customer_Cluster', 'Transaction_Amount'],
    'Phase 2 (Week 3-4)': ['Days_Since_Last_Transaction', 'Transaction_Frequency', 'Payment_Velocity'],
    'Phase 3 (Week 5-6)': ['Amount_Variability', 'Time_Consistency', 'Channel_Risk_Score'],
    'Phase 4 (Week 7-8)': ['Seasonal_Index', 'Product_Category', 'Time_of_Day']
}

# Create Gantt-style chart
fig_gantt, ax_gantt = plt.subplots(figsize=(12, 6))

phase_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
phase_durations = [2, 2, 2, 2]  # Weeks per phase
start_week = 0

for i, (phase, features) in enumerate(implementation_phases.items()):
    # Draw phase bar
    ax_gantt.barh(i, phase_durations[i], left=start_week, 
                  color=phase_colors[i], alpha=0.7, height=0.6)
    ax_gantt.text(start_week + phase_durations[i]/2, i, phase,
                 ha='center', va='center', color='white', fontweight='bold')
    
    # Add features
    for j, feature in enumerate(features):
        feature_y = i - 0.2 + j*0.15
        ax_gantt.text(start_week + 0.1, feature_y, f" {feature}",
                     ha='left', va='center', fontsize=8)
    
    start_week += phase_durations[i]

ax_gantt.set_yticks(range(len(implementation_phases)))
ax_gantt.set_yticklabels([p.split()[0] for p in implementation_phases.keys()])
ax_gantt.set_xlabel('Weeks from Start')
ax_gantt.set_title('Feature Implementation Timeline', fontsize=12, fontweight='bold')
ax_gantt.grid(axis='x', alpha=0.3)
ax_gantt.set_xlim(0, sum(phase_durations))

plt.tight_layout()
plt.savefig('../../reports/task2_eda/visualizations/summary/feature_timeline.png', 
            dpi=300, bbox_inches='tight')
plt.show()
```
